add column

```python
canada[years].sum(axis=1)           #total imigration for each country
```

```python
canada['Total'] = canada[years].sum(axis=1)         # a new column Total will be added to the end of the dataframe
```

```python
canada.head(2)
```

remove column

```python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
canada.drop(columns = cols_to_drop,inplace = True)
canada
```

```python
canada.drop_duplicates(subset=['AreaName'])         # just a example to removeb  duplicates
```

remaining columns

```python
canada.sample()
```

```python
canada.rename(mapper={'AreaName' : 'Continent',
                      'RegName' : 'Region',
                       'DevName' : 'Status'
                        },axis = 1 ,inplace = True)
```

```python
canada
```

```python
#reordering sequence
updated_list = ['Status','Continent','Region','Total']+years
print(updated_list)
```

```python
canada = canada[updated_list]
```

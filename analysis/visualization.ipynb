```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='',marker='o',markersize = 5,alpha=0.75,figsize = (15,7),markerfacecolor = 'yellow',markeredgecolor = 'black')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,5),color = 'green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,5),alpha= 0.5,color = 'green')
```

```python
penguins.plot(kind = 'scatter',x = 'bill_length_mm',y = 'bill_depth_mm',s = 100,alpha=0.5)
```

```python
penguins.plot(kind = 'scatter',x = 'bill_length_mm',y = 'flipper_length_mm',s = 100,alpha=0.5)
```

subplots

```python
fig, axs = plt.subplots(1,2, figsize = (15,5),constrained_layout = True)
penguins.plot(kind = 'scatter',x = 'bill_length_mm',y = 'flipper_length_mm',s = 100,alpha=0.5,ax=axs[0],title = 'penguin bill data')
penguins.plot(kind = 'scatter',x = 'bill_length_mm',y = 'bill_depth_mm',s = 100,alpha=0.5,ax = axs[1],title = 'penguin flipper')
plt.show()
```

```python
penguins['species'].value_counts()
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct='%1.1f%%',#add in percentage
                                        figsize = (3,3),
                                        title = 'Species',
                                        ylabel = '',
                                        pctdistance = .7,
                                        wedgeprops = dict(
                                            width = 0.5
                                        ))

                                    
```

# advance visualization

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data = tips,x = 'sex')
```

```python
sns.countplot(data = tips,x = 'sex',hue ='day')
```

```python
sns.displot(data = tips,x = 'total_bill',binwidth=5)                 # distribution plot (histogram)
```

```python
sns.displot(data = tips, x = 'total_bill',hue = 'sex',col ='sex')
```

```python
sns.displot(data = tips, x = 'total_bill',hue = 'smoker',row ='smoker')
```

```python
sns.displot(data = tips, x = 'total_bill',col = 'smoker',row ='sex',height = 3,kind = 'kde',hue = 'day',multiple = 'stack')
```
